```python
team_basic=pd.read_csv("team_basic.csv")
team_basic['name']=team_basic['name'].astype(str)
print(team_basic.head())
print(team_basic.shape[0],team_basic.shape[1])
```

```python
team_stat=pd.read_csv("team_stats.csv",index_col=False)
team_stat['Team']=team_stat['Team'].astype(str)
print(team_stat.head())
print(team_stat.shape[0],team_stat.shape[1])
```

```python
#Q2.1 Êï∞ÊçÆË°®ÂêàÂπ∂
#TeamË°®ÂêàÂπ∂
team=pd.merge(team_stat,team_basic,left_on='Team',right_on='name',how='left')

#ÈòüÂêç‰∏çÂêåÊ†ºÂºèÈóÆÈ¢ò
#Since the * indicates a playoff appearance, the team* should be changed to team.
team_basic['name'].nunique()
for i in range(team_stat.shape[0]):
    team_stat.loc[i,'Team']=team_stat.loc[i,'Team'].strip('*')
team_stat['Team'].unique()

team=pd.merge(team_stat,team_basic,left_on='Team',right_on='name',how='left')
print(team.head())
print(team.shape[0],team.shape[1])
```

```python
#PlayerË°®ÂêàÂπ∂
#Ë∞ÉÊï¥‰∏ªÈîÆÊ†ºÂºè‰∏ÄËá¥
for i in range(player_details.shape[0]):
    player_details.loc[i,'player_url']=player_details.loc[i,'player_url'][36:]
#print("salaries",player_salaries.head())
#print("stat",player_stat.head())
player=pd.merge(player_details,player_salaries,left_on=['player_url','Season'],right_on=['player_url','season'],how='left')
player=pd.merge(player,player_stat,left_on='player_url',right_on='url',how='left')
print(player.head())
print(player.shape[0],player.shape[1])
```

```python
#Later we extract the year information of season out and delete the duplicates "season" variables versus "Season" variables, and 
# and droped NaN of player dataframe, and got new_player.csv. 
player=pd.read_csv("new_player.csv")
player.head()
```

```python
#Clean the team dataframe, drop N/A records
team=pd.read_csv("team.csv", index_col=0)
team=team.dropna(axis=1)
team.head()
```

```python
#Q2 For the most current season in the dataset
#a how many active players?
player2=player[player['season_year']==2019]
player2=player2[player2['active']==True]
player2['player_url'].nunique()
#There are 386 active players in season 2019-2020
```

```python
#b how many players in each postion?
player2=player[player['season_year']==2019]
posplayer = player2['player_url'].groupby(player2['Pos'])
posplayer.count()
'''
Pos    Number
C        86
C,PF      5
C,SF      1
PF       75
PF,SF     5
PG       71
PG,SG     3
SF       65
SF,SG     3
SG       83
'''
```

```python
#c Average age, weight, experience, salary
player2=player[player['season_year']==2019]
print("Age",player2["Age"].mean())
print("Weight",player2['wt'].mean())
print("Experience",player2['experience(year)'].mean())
print("Salary",player2['salary'].mean())
```

```python
#d career salary
#everybody has different career, so group by player and compute mean salary
player2=player.groupby("player_url")
#then compute all players' career salary 
careers=player2['salary'].mean()
careers.mean()
```

```python
#3 a how many palyers were active in each season?
player2=player[player['active']==True]
player2=player2["player_url"].groupby(player2["season_year"])
player2.count()
'''
season_year
2009     65
2010     83
2011    106
2012    141
2013    160
2014    209
2015    248
2016    303
2017    362
2018    450
2019    386
'''
```

```python
#average salary by season
player2=player["salary"].groupby(player["season_year"])
player2.mean()
```

```python
#variance of salary be season
player2.var()
```

```python
#b top 10% paid palyers current season?
player2=player[player['season_year']==2019]
player2=player2.sort_values(axis=0,ascending=False,by=['salary'])
rank10=player2.shape[0]//10
print(player2.iloc[:rank10,:]["player_url"])
print(player2.iloc[:rank10,:]["Team"])
#We identify palyers by url, because there are players with the same name
#following are the best paid 39 athletes webs and their team
```

```python
#lowest 10% paid players
rank90=player2.shape[0]-rank10
print(player2.iloc[rank90:,:]["player_url"])
print(player2.iloc[rank90:,:]["Team"])
# Shaquille Harrison of CHI team is the lowest 39 paid players' urls and their team 
```

```python
#middle 50% paid players
rank50=player2.shape[0]//2
print(player2.iloc[rank50,:]["player_url"])
print(player2.iloc[rank50,:]["Team"])
#That guy is Bullore in team NYK
```

```python
#e total career payment for all current season actice players
player2=player[player['season_year']==2019]
player2=player2[player2["active"]==True]["player_url"]
totalpay=0
for i in range(player.shape[0]):
    if player.iloc[i,:]["player_url"] in list(player2):
        totalpay+=player.iloc[i,:]["salary"]
totalpay
```

```python
#4 team-player statistics
#a average salary of each team by season
player2=player["salary"].groupby([player["season_year"],player["Team"]])
player2.mean()
```

```python
#b average age of players by season
player2=player["Age"].groupby(player["season_year"])
player2.mean()
```

```python
#b average and variance of experience by season of each team
player2=player["experience(year)"].groupby([player["season_year"],player["Team"]])
player2.mean()
```

```python
#variance
player2.var()
```

```python
#c provide b's information in a tabulation format
import numpy as np
table=pd.pivot_table(player,index=["Team"],columns=["season_year"],values=["experience(year)"],aggfunc=[np.mean,np.var])
table
```
